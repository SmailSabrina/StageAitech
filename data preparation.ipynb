```python
import pandas as pd
data=pd.read_csv('data/data_set.csv')
```

```python
data.head(10)
```

##### la transformation de type d'angrais en trois colonnes(chaque type d'engrais dans un colonne separé)    

```python
data["Type d'engrais"].unique()
```

```python
data['Qte azote(N)'] = data["Type d'engrais"].apply(lambda x: 1 if x=="Élément nutritif azote N (total)" else 0)
data['Qte azote(N)']=data['Qte azote(N)']*data["Qte engrais(kg/ha)"]
#
data['Qte Engrais phosphatés'] = data["Type d'engrais"].apply(lambda x: 1 if x=="Engrais phosphatés (total d'élément fertilisant P205)" else 0)
data['Qte Engrais phosphatés']=data['Qte Engrais phosphatés']*data["Qte engrais(kg/ha)"]
#
data['Qte potasse K2O'] = data["Type d'engrais"].apply(lambda x: 1 if x=='Élément nutritif potasse K2O (total)' else 0)
data['Qte potasse K2O']=data['Qte potasse K2O']*data["Qte engrais(kg/ha)"]
#
data.head(5)
```

```python
new_df=data.groupby(['Zone','Année']).sum()
```

```python
new_df["Qte pesticide(kg/ha)"]=new_df["Qte pesticide(kg/ha)"]/3
new_df
```

```python
new_df.drop(["Qte engrais(kg/ha)"],axis=1,inplace=True)
```

```python
new_df.drop(["No_value"],axis=1,inplace=True)
```

```python
new_df.to_csv("data/new_data_set.csv")
```
