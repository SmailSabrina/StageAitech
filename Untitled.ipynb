### Production

```python
import pandas as pd
prod=pd.read_csv('data2/citrus production.csv')
prod.head()
```

```python
prod.drop(["Domain Code","Domain","Year Code","Element Code","Element","Item Code","Item","Flag","Flag Description","Unit"],axis=1,inplace=True)
```

```python
prod.rename(columns={
    'Value': 'Citrus Production(hg/ha)'},
    inplace=True)
prod
```

### Engrais

```python
eng=pd.read_csv('data2/engrais.csv')
eng.head()
```

```python
eng.drop(["Domain Code","Domain","Year Code","Element Code","Element","Item Code","Flag","Flag Description","Unit"],axis=1,inplace=True)
```

```python
eng
```

```python
eng["Item"].unique()
```

```python
eng['Qte azote(N)'] = eng["Item"].apply(lambda x: 1 if x=="Nutrient nitrogen N (total)" else 0)
eng['Qte azote(N)']=eng['Qte azote(N)']*eng["Value"]
#
eng['Qte Engrais phosphatÃ©s'] = eng["Item"].apply(lambda x: 1 if x=="Nutrient phosphate P2O5 (total)" else 0)
eng['Qte Engrais phosphatÃ©s']=eng['Qte Engrais phosphatÃ©s']*eng["Value"]
#
eng['Qte potasse K2O'] = eng["Item"].apply(lambda x: 1 if x=='Nutrient potash K2O (total)' else 0)
eng['Qte potasse K2O']=eng['Qte potasse K2O']*eng["Value"]
#
eng.head(5)
```

```python
eng=eng.groupby(['Area','Year']).sum()
```

```python
eng.drop(["Value"],axis=1,inplace=True)
eng
```

### temperature

```python
temp=pd.read_csv('data2/temp change.csv')
temp.head()
```

```python
temp.drop(["Domain Code","Domain","Year Code","Element Code","Element","Months Code","Flag","Flag Description","Unit"],axis=1,inplace=True)
```

```python
temp
```

```python
temp['Months'].unique()
```

```python
temp['temp Marâ€“Aprâ€“May'] = temp['Months'].apply(lambda x: 1 if x=='Marâ€“Aprâ€“May' else 0)
temp['temp Marâ€“Aprâ€“May']=temp['temp Marâ€“Aprâ€“May']*temp["Value"]
#
temp['temp Junâ€“Julâ€“Aug'] = temp['Months'].apply(lambda x: 1 if x=='Junâ€“Julâ€“Aug' else 0)
temp['temp Junâ€“Julâ€“Aug']=temp['temp Junâ€“Julâ€“Aug']*temp["Value"]
#
temp['temp Sepâ€“Octâ€“Nov'] = temp['Months'].apply(lambda x: 1 if x=='Sepâ€“Octâ€“Nov' else 0)
temp['temp Sepâ€“Octâ€“Nov']=temp['temp Sepâ€“Octâ€“Nov']*temp["Value"]
#
temp['temp year'] = temp['Months'].apply(lambda x: 1 if x=='Meteorological year' else 0)
temp['temp year']=temp['temp year']*temp["Value"]
temp
```

```python
temp=temp.groupby(['Area','Year']).sum()
```

```python
temp.drop(["Value"],axis=1,inplace=True)
temp
```

### Join them all

```python
T1=pd.merge(prod,pes,how='left',left_on=['Area','Year'],right_on=['Area','Year'])
T1
```

```python
T2=pd.merge(T1,eng,how='left',left_on=['Area','Year'],right_on=['Area','Year'])
T2.drop(["Area Code_x","Area Code_y","Area Code"],axis=1,inplace=True)
T2
```

```python
fin=pd.merge(T2,temp,how='left',left_on=['Area','Year'],right_on=['Area','Year'])
fin.drop(["Area Code"],axis=1,inplace=True)
fin
```

```python
fin.to_csv("data2/citrus prod dt.csv")
```
